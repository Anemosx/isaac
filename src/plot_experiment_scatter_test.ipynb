```python

for runs, title, split_run in zip(all_runs, titles, split_runs):
    y_range = [-0.01, 1.01]
    x_range = [-0.01, 1.01]



    epoch_return = None
    df = None

    with open('neptune_auth.json', 'r') as f:
        neptune_auth = json.load(f)
        neptune_auth = DotMap(neptune_auth)

    def get_df_from_neptune(run, run_id, split):
        neptune_run = neptune.init(project=neptune_auth.project, api_token=neptune_auth.api_token, mode="read-only", run="AA-{}".format(run_id))
        if split and use_test:
            epoch_return = neptune_run["win_test"].fetch_values(include_timestamp=False)['value']
        else:
            epoch_return = neptune_run["win"].fetch_values(include_timestamp=False)['value']
        df = pd.DataFrame()
        df["x"] = range(1, len(epoch_return) + 1)
        df["win"] = epoch_return.rolling(100).mean()
        try:
            if split and use_test:
                epoch_same = neptune_run["ag_0_same_a_p_test"].fetch_values(include_timestamp=False)['value']
            else:
                epoch_same = neptune_run["ag_0_same_a_p"].fetch_values(include_timestamp=False)['value']

            if run in algos:
                df["same"] = 1.
            else:
                df["same"] = epoch_same.rolling(100).mean()
        except:
            df["same"] = 1.

        neptune_run.stop()
        return df

    og_dataframes = []

    for run, split in zip(runs, split_run):
        dataframe = []
        for nep_id in run:
            dataframe.append(get_df_from_neptune(run, nep_id, split))
        dataframe = pd.concat(dataframe, ignore_index=True)
        dataframe.sort_values('x', inplace=True)

        og_dataframes.append(dataframe)

    current_palette = sns.color_palette(flatui)

    colors_purples = sns.color_palette("Purples", as_cmap=True)
    colors_reds = sns.color_palette("Reds", as_cmap=True)
    colors_greys = sns.color_palette("Greys", as_cmap=True)
    colors_greens = sns.color_palette("Greens", as_cmap=True)
    colors_blues = sns.color_palette("Blues", as_cmap=True)

    palettes = [colors_purples, colors_blues, colors_greens, colors_reds]

    markers = ["s", "o", "P", "X"]

    norm = plt1.Normalize(0, length)
    sm = plt1.cm.ScalarMappable(cmap = colors_greys, norm=norm)
    sm.set_array([])


    plt = None
    for i, dataframe in enumerate(og_dataframes):
        plt = sns.scatterplot(x="win", y="same", data=dataframe, hue="x", color=current_palette[i], linewidth=0,palette=palettes[i], marker=markers[i], rasterized=True)

    plt1.gca().set_aspect('equal', adjustable='box')
    plt.set_xlabel("win %", fontsize=20)
    plt.set_ylabel("same action %", fontsize=20)

    colors = []
    if len(labels) == 0:
        labels = runs
    for i, label in enumerate(labels):
        colors.append(mpatches.Patch(color=current_palette[i], label='{}'.format(label)))

    # if add_stuff:
    #     cb = plt.figure.colorbar(sm)
    #     cb.set_label('episode')
    cb = plt.figure.colorbar(sm)
    cb.ax.tick_params(labelsize=20)
    cb.set_label('episode')

    legend_elements = [Line2D([0], [0], marker='s', lw=4, color='w', markerfacecolor=mpl.colors.to_hex(colors_purples(0.7)), markersize=10, label=labels[0]),
                       Line2D([0], [0], marker='o', lw=4, color='w', markerfacecolor=mpl.colors.to_hex(colors_blues(0.7)), markersize=10, label=labels[1]),
                       Line2D([0], [0], marker='P', lw=4, color='w', markerfacecolor=mpl.colors.to_hex(colors_greens(0.7)), markersize=10, label=labels[2]),
                       Line2D([0], [0], marker='X', lw=4, color='w', markerfacecolor=mpl.colors.to_hex(colors_reds(0.7)), markersize=10, label=labels[3]),]

    leg = plt.legend(handles=legend_elements, labelspacing=0.1, loc=legend_location)
    leg.set_title(title)

    plt.set(xlim=(x_range[0], x_range[1]))
    plt.set(ylim=(y_range[1], y_range[0]))

    plt.set_xticks(np.arange(0, 1.01, step=0.5))
    plt.set_yticks(np.arange(0, 1.01, step=0.5))


    plt1.xticks(fontsize=20)
    plt1.yticks(fontsize=20)

    # if not add_stuff:
    #     plt.legend([],[], frameon=False)
    #     plt.tick_params(bottom=False, left=False)
    #     plt1.tick_params(bottom=False, left=False)





    fig = plt.get_figure()
    points = fig.axes[0].collections[0]
    points.set_rasterized(True)

    timestamp = datetime.datetime.now().strftime("%Y_%m_%d_%H_%M_%S")

    plot_dir = os.path.join(os.getcwd(), 'plots')
    if not os.path.exists(plot_dir):
        os.makedirs(plot_dir)
    fig.savefig(os.path.join(plot_dir, 'plot - {}.svg'.format(timestamp)), dpi=2000, bbox_inches="tight", format="svg")
    fig.savefig(os.path.join(plot_dir, 'plot - {}.png'.format(timestamp)), dpi=1000, bbox_inches="tight")
    # fig.savefig(os.path.join(plot_dir, 'plot - {}.eps'.format(timestamp)), dpi=2000, bbox_inches="tight", format='eps')
    # plot_as_emf(fig, save_path=plot_dir, save_name='plot - {}'.format(timestamp))
    fig.clf()
```
