```python
runs = []
if agent_type == "No Substitution":
    runs = algo_runs
elif agent_type == "Hidden Attacker":
    runs = h_runs
elif agent_type == "Obvious Attacker":
    runs = obv_runs
elif agent_type == "Random Agent":
    runs = random_runs

def m_ci(data, confidence=0.95):
    h = scipy.stats.sem(data) * scipy.stats.t.ppf((1 + confidence) / 2., len(data)-1)
    return h

with open('neptune_auth.json', 'r') as f:
    neptune_auth = json.load(f)
    neptune_auth = DotMap(neptune_auth)

neptune_run_data = []

for run in runs:
    data = []
    for nep_id in run:
        neptune_run = neptune.init(project=neptune_auth.project, api_token=neptune_auth.api_token, mode="read-only", run="AA-{}".format(nep_id))

        try:
            data_same = neptune_run["ag_0_same_a_p"].fetch_values(include_timestamp=False)['value']
        except:
            data_same = None

        data.append(DotMap(
            returns=neptune_run["return"].fetch_values(include_timestamp=False)['value'],
            same_actions=data_same,
            ally_healths=neptune_run["ally_health"].fetch_values(include_timestamp=False)['value'],
            enemy_healths=neptune_run["enemy_health"].fetch_values(include_timestamp=False)['value'],
            wins=neptune_run["win"].fetch_values(include_timestamp=False)['value'],
        ))
        neptune_run.stop()
    neptune_run_data.append(data)
```

```python
print(agent_type, "\n")
for run_data, label in zip(neptune_run_data, labels):
    returns = []
    same_actions = []
    ally_healths = []
    enemy_healths = []
    wins = []
    for run_d in run_data:
        returns = [*returns, *run_d.returns.to_list()[:1000]]
        ally_healths = [*ally_healths, *run_d.ally_healths.to_list()[:1000]]
        enemy_healths = [*enemy_healths, *run_d.enemy_healths.to_list()[:1000]]
        wins = [*wins, *run_d.wins.to_list()[:1000]]
        try:
            same_actions = [*same_actions, *run_d.same_actions.to_list()[:1000]]
        except:
            same_actions = [1.0, 1.0]

    returns = np.array(returns)
    same_actions = np.array(same_actions)
    ally_healths = np.array(ally_healths)
    enemy_healths = np.array(enemy_healths)
    wins = np.array(wins)

    print("===============================================")
    print(f"{label}\t\t\t\t\t\t\t       {environment}")
    print("_______________________________________________")
    print(f"wins            | mean+-ci:\t  ${round(wins.mean(), 2):.2f} \pm {round(m_ci(wins), 2):.2f}$")
    print(f"same_actions    | mean+-ci:\t  ${round(same_actions.mean(), 2):.2f} \pm {round(m_ci(same_actions), 2):.2f}$")
    print(f"return          | mean+-ci:\t  ${round(returns.mean(), 2):.2f} \pm {round(m_ci(returns), 2):.2f}$")
    print(f"ally_healths    | mean+-ci:\t  ${round(ally_healths.mean(), 2):.2f} \pm {round(m_ci(ally_healths), 2):.2f}$")
    print(f"enemy_healths   | mean+-ci:\t  ${round(enemy_healths.mean(), 2):.2f} \pm {round(m_ci(enemy_healths), 2):.2f}$")
    print("===============================================\n\n")


    # print("===============================================")
    # print(f"{label}\t\t\t\t\t\t\t       {environment}")
    # print("_______________________________________________")
    # print(f"return           | mean: {round(returns.mean(), 2):.2f}\t\t| std: {round(returns.std(), 2):.2f}")
    # print(f"same_actions     | mean: {round(same_actions.mean(), 2):.2f}\t\t| std: {round(same_actions.std(), 2):.2f}")
    # print(f"ally_healths     | mean: {round(ally_healths.mean(), 2):.2f}\t\t| std: {round(ally_healths.std(), 2):.2f}")
    # print(f"enemy_healths    | mean: {round(enemy_healths.mean(), 2):.2f}\t\t| std: {round(enemy_healths.std(), 2):.2f}")
    # print(f"wins             | mean: {round(wins.mean(), 2):.2f}\t\t| std: {round(wins.std(), 2):.2f}")
    # print("===============================================\n\n")
```
