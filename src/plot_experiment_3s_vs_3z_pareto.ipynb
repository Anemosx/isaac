```python
for runs, title in zip(all_runs, titles):

    # runs = [runs[3], runs[1], runs[2], runs[0]]

    y_range = [-0.01, 1.01]
    x_range = [-0.01, 1.01]

    epoch_return = None
    df = None

    with open('neptune_auth.json', 'r') as f:
        neptune_auth = json.load(f)
        neptune_auth = DotMap(neptune_auth)

    def get_df_from_neptune(run_id):
        neptune_run = neptune.init(project=neptune_auth.project, api_token=neptune_auth.api_token, mode="read-only", run="AA-{}".format(run_id))
        epoch_return = neptune_run["win"].fetch_values(include_timestamp=False)['value']
        df = pd.DataFrame()
        df["x"] = range(1, len(epoch_return) + 1)
        df["win"] = epoch_return.rolling(100).mean()
        try:
            epoch_same = neptune_run["ag_0_same_a_p"].fetch_values(include_timestamp=False)['value']
            df["same"] = epoch_same.rolling(100).mean()
        except:
            df["same"] = 1.

        neptune_run.stop()
        return df

    dataframes = []
    for run in runs:
        dataframe = []
        for nep_id in run:
            dataframe.append(get_df_from_neptune(nep_id))
        dataframe = pd.concat(dataframe, ignore_index=True)

        dataframe = dataframe.groupby('x', as_index=False)[['win', 'same']].mean()

        dataframes.append(dataframe)

    current_palette = sns.color_palette(flatui)

    colors_purples = sns.color_palette("Purples", as_cmap=True)
    colors_reds = sns.color_palette("Reds", as_cmap=True)
    colors_greys = sns.color_palette("Greys", as_cmap=True)
    colors_blues = sns.color_palette("Blues", as_cmap=True)

    palettes = [colors_reds, colors_blues, colors_greys, colors_purples]

    plt = None
    for i, dataframe in enumerate(dataframes):
        # plt = sns.lineplot(x="win", y="same", hue="x", data=dataframe, color=current_palette[i],palette=palettes[i],  err_style='band', ci="sd")

        plt = sns.lineplot(x="win", y="same", data=dataframe, color=current_palette[i], err_style='band', ci="sd")

    plt.axhline(0.4, linestyle="--", lw=1, color="turquoise", alpha=0.4)
    plt.axhline(0.5, linestyle="--", lw=1, color="#2ecc71", alpha=0.4)
    plt.set_xlabel("win")
    plt.set_ylabel("same action")

    colors = []
    if len(labels) == 0:
        labels = runs
    for i, label in enumerate(labels):
        colors.append(mpatches.Patch(color=current_palette[i], label='{}'.format(label)))

    leg = plt.legend(handles=colors, labelspacing=0.1, loc=legend_location)
    leg.set_title(title)

    plt.set(xlim=(x_range[0], x_range[1]))
    plt.set(ylim=(y_range[1], y_range[0]))

    fig = plt.get_figure()

    timestamp = datetime.datetime.now().strftime("%Y_%m_%d_%H_%M_%S")

    plot_dir = os.path.join(os.getcwd(), 'plots')
    if not os.path.exists(plot_dir):
        os.makedirs(plot_dir)
    fig.savefig(os.path.join(plot_dir, 'plot - {}.png'.format(timestamp)), dpi=300, bbox_inches="tight")
    fig.clf()
```
